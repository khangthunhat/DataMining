```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
means = [[2, 2], [9, 2], [4, 9], [1, 8], [8, 7]]
cov = [[2, 0], [0, 2]]
n_samples = 500
n_cluster = 3
X0 = np.random.multivariate_normal(means[0], cov, n_samples)
X1 = np.random.multivariate_normal(means[1], cov, n_samples)
X2 = np.random.multivariate_normal(means[2], cov, n_samples)
X3 = np.random.multivariate_normal(means[3], cov, n_samples)
X4 = np.random.multivariate_normal(means[4], cov, n_samples)
X = np.concatenate((X0, X1, X2, X3, X4), axis = 0)
```

```python
plt.xlabel('x')
plt.ylabel('y')
plt.plot(X[:, 0], X[:, 1], 'bo', markersize=5)
plt.plot()
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D # 3d plot
from termcolor import colored as cl # text customization

from sklearn.preprocessing import StandardScaler # data normalization
from sklearn.cluster import KMeans # K-means algorithm
clusters = 2
model = KMeans(
    n_clusters = clusters)
model.fit(X)

labels = model.labels_
print(cl(labels[:100], attrs = ['bold']))

```

```python
model = KMeans(init = 'k-means++',
               n_clusters = clusters,
               n_init = 12)
```

```python
model.fit(X)

labels = model.labels_
print(cl(labels[:100], attrs = ['bold']))
```

```python
kmeans = KMeans(n_clusters=3, random_state=42).fit(X)
```

```python
cluster_centers = kmeans.cluster_centers_
labels = kmeans.labels_
```

```python
plt.scatter(X[:, 0], X[:, 1], c=labels)
plt.scatter(cluster_centers[:, 0], cluster_centers[:, 1], marker='x', color='red')
plt.show()
```
