```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules

# Đọc dữ liệu từ file
data = pd.read_csv('Market_Basket_AR.csv')

data
```

```python

# Chuyển đổi dữ liệu thành định dạng phù hợp
te = TransactionEncoder()
te_arr = te.fit_transform(data)
df = pd.DataFrame(te_arr, columns=te.columns_)
```

```python
df
```

```python
# Tìm frequent itemsets với mức độ hỗ trợ 0.01
frequent_itemsets_001 = apriori(df, min_support=0.005, use_colnames=True)
```

```python
rules = association_rules(frequent_itemsets_001, metric="confidence", min_threshold=0.3)
```
