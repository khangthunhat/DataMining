$$ x_{norm} = \frac{x - x_{min}}{x_{max} - x_{min}} $$

trong đó:

$x$: giá trị ban đầu của một phần tử trong cột.
$x_{norm}$: giá trị chuẩn hóa của phần tử đó.
$x_{min}$: giá trị nhỏ nhất của cột.
$x_{max}$: giá trị lớn nhất của cột.

```python
data['youtube_norm'] = (data['youtube'] - data['youtube'].min()) / (data['youtube'].max() - data['youtube'].min())
data['facebook_norm'] = (data['facebook'] - data['facebook'].min()) / (data['facebook'].max() - data['facebook'].min())
data['newspaper_norm'] = (data['newspaper'] - data['newspaper'].min()) / (data['newspaper'].max() - data['newspaper'].min())
data['sales_norm'] = (data['sales'] - data['sales'].min()) / (data['sales'].max() - data['sales'].min())
```
