```python
import pandas as pd 
import numpy as np

# Đọc file dữ liệu vào dataFrame
data = pd.read_csv('ThiTHPT2018.csv')
data
```

```python
data[data.isnull().any(axis=1)].head()
```

```python
# show null values of col
data.isnull().sum()
```

```python
data[['Toan', 'Van', 'Anh']].describe()
```

```python
# Giá trị xuất hiện nhiều nhất của ba môn Toán Văn Anh
data[['Toan', 'Van', 'Anh']].mode()
```

```python
data1 = data.copy()
# Replacing null value with mean of Toan Attribute
data1['Toan'].fillna(data1['Toan'].mean(), inplace= True)
```

```python
# Replacing null value with mode of Van Attribute
data1['Van'].fillna(data1['Van'].mode()[0], inplace=True)
```

```python
data1.isnull().sum()
```

```python
# Gía trị xuất hiện nhiều nhất của môn Anh theo từng tỉnh
mode_values = data1.groupby('Ten Tinh')['Anh'].apply(lambda x: x.mode())
print(mode_values)
```

```python
# Kết quả của toàn bộ đoạn mã này là cập nhật lại cột 'Anh' trong DataFrame data1 với các giá trị bị thiếu được điền 
# bằng giá trị mode của nhóm tương ứng theo giá trị trong cột 'Ten Tinh'.
data1['Anh'] = data1.groupby(['Ten Tinh'], group_keys=False)['Anh'].apply(lambda x: x.fillna(x.mode()))
```

```python
data1.isnull().sum()
```

```python
# Thay thế những giá trị thiếu bằng 'Không Thi'
data1[['Toan', 'Van', 'Anh', 'Ly', 'Hoa', 'Sinh', 'Su', 'Dia', 'GDCD']].fillna('Không thi')
```

```python

# tính số giá trị thiếu cho từng cột
missing_values = data1.isnull().sum()

# tính tổng số giá trị trong DataFrame
total_values = np.product(data1.shape)

# tính tỷ lệ giá trị thiếu cho từng cột
missing_ratio = missing_values / total_values

# xác định các cột có tỷ lệ giá trị thiếu >= 0.5
cols_to_drop = missing_ratio[missing_ratio >= 0.5].index.tolist()

# xóa các cột có tỷ lệ giá trị thiếu >= 0.5
data1 = data1.drop(cols_to_drop, axis=1)
```

```python
#xoá trường không có dữ liệu
data1.dropna(axis=1, how='all',inplace=True)
```

```python
data1.isnull().sum()
```

```python
#điền giá trị '0' cho những ô bị thiếu
data1.fillna('0',inplace=True)
```

```python
#điền giá trị  NaN cho gía trị bị thiếu
data1 = data1.fillna('NaN')
```

```python
data1.dropna(subset=['Hoa', 'Ly', 'Dia'], inplace=True)
```

```python
data1.to_csv('TotnghiepOK.csv', index=False)
```
