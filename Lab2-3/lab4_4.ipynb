```python
import pandas as pd

data = pd.read_csv("./105.csv")

data
```

Rời rạc hóa các trường trong tệp dữ liệu điểm 105.csv thành 0,1 (0 cho những
giá trị điểm >=6, 0 cho những giá trị còn lại) Lưu lại thành file 105moi.csv

```python
data_col = data.columns.tolist()
print(data_col)
```

```python
# Xoá cột Unnamed: 55
data = data.drop(columns=['Unnamed: 55'])

# Định nghĩa hàm lambda để rời rạc hóa các giá trị điểm
f = lambda x: 1 if x >=6 else 0

# Rời rạc hóa tất cả các trường dữ liệu bằng phương thức applymap
data_roirac = data.applymap(f)

data_roirac.to_csv('105moi.csv', index=False)
```
