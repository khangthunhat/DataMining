```python
import pandas as pd

# Đọc file dữ liệu 105.csv
data = pd.read_csv('105.csv')

# Sử dụng hàm apply() để chuyển đổi từng cột về dạng true, false
data_tf = data.apply(lambda x: ~x.isna())

# Lưu kết quả vào file 105_tf.csv
data_tf.to_csv('105_tf.csv', index=False)
```

Thực hiện chuyển dổi cho file dữ liệu 105.csv về dạng true, false. (true
cho những cột có điểm, false cho những cột không có điểm)
