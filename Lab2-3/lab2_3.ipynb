```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_excel('./CollegeRecruitingData.xlsx')

data
```


Years of Higher Education: Thông tin về trình độ học vấn của nhân viên có thể liên quan đến khả năng đóng góp và sự phát triển trong công việc.

Years of Experience: Thông tin về số năm kinh nghiệm làm việc có thể giúp đánh giá khả năng làm việc hiệu quả của nhân viên.

Age: Thông tin về độ tuổi có thể giúp đánh giá khả năng và kinh nghiệm của nhân viên trong công việc.

Recruitment Bonus: Thông tin về mức độ hấp dẫn của chương trình tuyển dụng và các chính sách thưởng có thể ảnh hưởng đến số lượng ứng viên và hiệu quả tuyển dụng.

Calls per Day và Emails per Day: Thông tin về số lượng cuộc gọi và email được thực hiện mỗi ngày có thể cho thấy khả năng liên lạc và quản lý thời gian của nhân viên.

Recruits per Month: Thông tin về số lượng nhân viên mới được tuyển dụng trong tháng có thể giúp đánh giá hiệu quả của quy trình tuyển dụng và năng suất làm việc của nhân viên.

```python

# Select the columns of interest
years_of_experience = data['Years of Experience']
years_of_education = data['Years of Higher Education']

# Create a scatter plot to show the relationship between the two variables
plt.scatter(years_of_experience, years_of_education)
plt.xlabel('Years of Experience')
plt.ylabel('Years of Higher Education')
plt.title('Relationship between Years of Experience and Years of Higher Education')
plt.show()
```

```python
# Select the columns of interest
years_of_experience = data['Years of Experience']
recruitment_bonus = data['Recruitment Bonus']

# Create a scatter plot to show the relationship between the two variables
plt.scatter(years_of_experience, recruitment_bonus)
plt.xlabel('Years of Experience')
plt.ylabel('Recruitment Bonus')
plt.title('Relationship between Years of Experience and Recruitment Bonus')
plt.show()
```

```python
import seaborn as sns

# violin plot phân tích sự phân bố Số năm học đại học
sns.violinplot(x=data['Years of Higher Education'])
plt.xlabel('Years of Higher Education')
plt.title('Distribution of Years of Higher Education')
plt.show()

# line chart phân tích xu hướng Cuộc gọi mỗi ngày theo thời gian
sns.lineplot(x=data.index, y=data['Calls per Day'])
plt.xlabel('Time')
plt.ylabel('Calls per Day')
plt.title('Trend of Calls per Day over Time')
plt.show()
```

```python


# Create histogram to analyze the distribution of Age
plt.hist(data['Age'], bins=20)
plt.xlabel('Age')
plt.ylabel('Count')
plt.title('Distribution of Age')
plt.show()

# Create box plot to analyze the distribution of Recruitment Bonus
plt.boxplot(data['Recruitment Bonus'])
plt.xlabel('Recruitment Bonus')
plt.title('Distribution of Recruitment Bonus')
plt.show()
```
