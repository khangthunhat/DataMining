```python
import pandas
import seaborn
import matplotlib.pyplot as plt
data_table = pandas.read_csv('ThiTHPT2018.csv')

data_table
```

```python
# Thống kê dữ liệu
data_table.describe().round(2)
```

```python
# Điểm trung bình các môn
print("Diem trung binh:")
data_table[['Toan', 'Van', 'Anh']].mean().round(2)
```

```python
# Giá trị thường xuyên của các môn thi
print('Gia tri thuong xuyen cac mon')
data_table[['Toan', 'Van', 'Anh', 'Ly', 'Hoa', 'Sinh', 'Su', 'Dia', 'GDCD']].mean().round(2)
```

```python
# Tứ vị phân các môn
print('\nTứ vị phân thứ nhất Q1:')
print(data_table[['Toan', 'Van', 'Anh']].quantile(.25))
print('\nTứ vị phân thứ hai Q2:')
print(data_table[['Toan', 'Van', 'Anh']].quantile(.5))
print('\nTứ vị phân thứ ba Q3:')
print(data_table[['Toan', 'Van', 'Anh']].quantile(.75))
```

```python
# Thống kê Toán Văn Anh theo từng tỉnh
data_table.groupby('Ten Tinh')[['Toan','Van','Anh']].describe().round(2)
```

```python
# Thống kê các môn học bằng biểu đồ Boxplot
seaborn.boxplot(data_table[['Toan', 'Van', 'Anh', 'Ly', 'Hoa', 'Sinh', 'Su', 'Dia', 'GDCD']])
```

```python
seaborn.heatmap(data_table[['Toan', 'Van', 'Anh']].corr(), annot=True).set(title='Tương quan giữa ba môn Toán, Văn, Anh')
plt.show()

seaborn.heatmap(data_table[['Ly', 'Hoa', 'Sinh']].corr(), annot=True).set(title='Tương quan giữa Tổ hợp môn tự nhiên')
plt.show()

seaborn.heatmap(data_table[['Su', 'Dia', 'GDCD']].corr(), annot=True).set(title='Tương quan giữa Tổ hợp môn xã hội')
plt.show()
```

```python
seaborn.histplot(x=data_table['Toan']).set(title='Điểm toán')
plt.show()
seaborn.histplot(x=data_table['Van']).set(title='Điểm Văn')
plt.show()
seaborn.histplot(x=data_table['Anh']).set(title='Điểm Anh')
plt.show()
seaborn.histplot(x=data_table['Ly']).set(title='Điểm Lý')
plt.show()
seaborn.histplot(x=data_table['Hoa']).set(title='Điểm Hóa')
plt.show()
seaborn.histplot(x=data_table['Sinh']).set(title='Điểm Sinh')
plt.show()
seaborn.histplot(x=data_table['Su']).set(title='Điểm Sử')
plt.show()
seaborn.histplot(x=data_table['Dia']).set(title='Điểm Địa')
plt.show()
seaborn.histplot(x=data_table['GDCD']).set(title='Điểm GDCD')
plt.show()
```

```python
# Thống kê các môn học

plt.figure(figsize=(10,5))
seaborn.boxplot(data=data_table[['Toan', 'Van', 'Anh', 'Ly', 'Hoa', 'Sinh', 'Su', 'Dia', 'GDCD']])
plt.title('Thống kê các môn học', fontsize=16)
plt.xlabel('Môn học', fontsize=14)
plt.ylabel('Điểm', fontsize=14)
plt.show()
```
