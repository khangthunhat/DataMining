```python
import pandas
import seaborn as sns
import numpy
import matplotlib.pyplot as plt

data = pandas.read_excel('JOB_INFO.xlsx')
data
```

```python
# Thống kê dữ liệu
data.describe().round(2)
```

```python
# Đo độ tập trung, phân tán và tóm tắt dữ liệu
# data.plot(kind='box')
sns.boxplot(x=data['Gender'], y=data['Salary'])
```

```python
# sns.set()
sns.histplot(data['Age'], bins=numpy.arange(data['Age'].min(), data['Age'].max(), 1)).set(title="Thống kê cột Age")
plt.ylabel("Số lượng")
plt.show()

sns.histplot(data['Dept']).set(title="Thống kê cột Dept")
plt.ylabel("Số lượng")
plt.show()

sns.histplot(data['Gender']).set(title="Thống kê cột Gender")
plt.ylabel("Số lượng")
plt.show()

sns.histplot(data['Skill']).set(title="Thống kê cột Skill")
plt.ylabel("Số lượng")
plt.show()

sns.histplot(data['WorkingYear'], bins=numpy.arange(data['WorkingYear'].min(), data['WorkingYear'].max(), 1)).set(title="Thống kê cột WorkingYear")
plt.ylabel("Số lượng")
plt.show()

sns.histplot(data['Salary'], bins=numpy.arange(data['Salary'].min(), data['Salary'].max(), 100)).set(title="Thống kê cột Salary")
plt.ylabel("Số lượng")
plt.show()
```

```python
# Đo độ tương quan dữ liệu 
sns.scatterplot(data=data, x='WorkingYear', y='Salary').set(title='Sự tương quan của trường Salarry và WorkingYear')
plt.show()

sns.scatterplot(data=data, x='Age', y='Salary', hue='Dept', style='Dept').set(title='Sự tương quan của trường Salarry và Age')
plt.show()
```

```python
sns.heatmap(data[['Age','Salary','TrainedYear', 'WorkingYear']].corr(), annot=True)

plt.show()
```
