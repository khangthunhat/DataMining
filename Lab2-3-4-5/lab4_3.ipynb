```python
import pandas

data = pandas.read_csv('./TotnghiepOK.csv')
```

```python

data1 = data.copy()
data2 = data.copy()
data
```

```python
def convert_score_to_category(score):
    bins = [0, 5, 6.5, 8, 10]
    categories = ['Yếu', 'Trung bình', 'Khá', 'Giỏi']
    return pandas.cut(score, bins=bins, labels=categories, include_lowest=True)

data1[['Toan', 'Van', 'Anh', 'Ly', 'Hoa', 'Sinh', 'Su', 'Dia', 'GDCD']] = data1[['Toan', 'Van', 'Anh', 'Ly', 'Hoa', 'Sinh', 'Su', 'Dia', 'GDCD']].apply(convert_score_to_category)

data1
```

```python
def convert_score_to_category(score):
    if score < 6.5:
        return 'TB'
    elif score < 8:
        return 'Khá'
    else:
        return 'Giỏi'

# Áp dụng hàm vào cột Anh
data2['Anh'] = data2['Anh'].apply(convert_score_to_category)
```
