```python
import matplotlib.pyplot as plt
import pandas as pd

# Tạo DataFrame
Data = pd.DataFrame({
   'Toan': [8, 7, 10, 9, 8],
   'Van': [7, 8, 9, 10, 8],
   'Anh': [9, 8, 7, 10, 7]
})

# Tính toán số lượng học sinh đạt được từng điểm số
value_counts_Toan = Data['Toan'].value_counts()
value_counts_Van = Data['Van'].value_counts()
value_counts_Anh = Data['Anh'].value_counts()

# Vẽ biểu đồ cột
plt.bar(['Toan', 'Van', 'Anh'], [value_counts_Toan[8], value_counts_Van[8], value_counts_Anh[8]])

# Thêm tiêu đề và các nhãn trục
plt.title('Số lượng học sinh đạt được điểm số 8')
plt.xlabel('Môn học')
plt.ylabel('Số lượng học sinh')

# Hiển thị biểu đồ
plt.show()



```
